f(x) =0.6x^3+5.5x^2+10x -5
1. Определить корни 
2. Найти интервалы, на которых функция возрастает 
3. Найти интервалы, на которых функция убывает
4. Построить график 
5. Вычислить вершины 
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
a = 0.6
b = 5.5
c = 10
d = 5
y = 0.6*x**3 + 5.5*x**2 + 10*x - 5
y
```

# График функции

```python
plot(y, (x, -8, 2.5), legend= True)
y
```

# Корни

```python
qw = solve(y, x)
```

```python
i = 1
for _ in qw:
    print(f'Корень {i} = {round(_, 2)}')
    i +=1
```

```python
der = diff(y)
der
```

```python
plot(y, der, (x, -8, 2.5), legend= True)
y, der
```

# Вершины функции

```python
q1 = solve(der, x)[0].n(3)
q2 = solve(der, x)[1].n(3)
q1, q2
```

```python
q1= solve(der, x)[0].n(3)
ext1 = 0.6*q1**3 + 5.5*q1**2 + 10*q1 - 5
print(f'Первая вершина функции {y} находится в точке "Х" = {q1} и "Y" равен {round(ext1, 2)}')
```

```python
q2= solve(der, x)[1].n(3)
ext2 = 0.6*q2**3 + 5.5*q2**2 + 10*q2 - 5
print(f'Вторая вершина функции {y} находится в точке "Х" = {q2} и "Y" равен {round(ext2, 2)}')
```

# F(x) - возрастает

```python
solve(der > 0)
```

```python
print(f'Функция возрастает {solve(der > 0)}')
```

# F(x) - убывает

```python
solve(der < 0)
```

```python
print(f'Функция убывает {solve(der < 0)}')
```

# F(x) > 0

```python
solve(y > 0)
```

```python
print(f'Функция больше 0 {solve(y > 0)}')
```

# F(x) < 0

```python
solve(y < 0)
```

```python
print(f'Функция меньше 0 {solve(y < 0)}')
```
